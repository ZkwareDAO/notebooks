# 零知识证明应用：爱因斯坦推理题转换为多项式

来看看这道爱因斯坦出的测试题, 他说世界上有98%的人回答不出,<br>
谁养鱼？<br>
1、在一条街上，有5座房子，喷了5种颜色。<br>
2、每个房子里住着不同国籍的人。<br>
3、每个人喝着不同的饮料，抽不同品牌的香烟，养不同的宠物。<br>
问题是：谁养鱼？<br>
提示：
1、英国人住红色房子。<br>
2、瑞典人养狗。<br>
3、丹麦人喝茶。<br>
4、绿色房子在白色房子左面。<br>
5、绿色房子主人喝咖啡。<br>
6、抽Pall Mall香烟的人养鸟。<br>
7、黄色房子主人抽Dunhill香烟。<br>
8、住在中间房子的人喝牛奶。<br>
9、挪威人住第一间房。<br>
10、抽Blends香烟的人住在养猫的人隔壁。<br>
11、养马的人住抽Dunhill香烟的人隔壁。<br>
12、抽Blue Master的人喝啤酒。<br>
13、德国人抽Prince香烟。<br>
14、挪威人住蓝色房子隔壁。<br>
15、抽Blends香烟的人有一个喝水的邻居。<br>

咋一看这个问题是一个逻辑推理题，根本不是计算类问题，把它转换成计算类问题，这个需要一定的想象力。我居然是在梦中想到的办法，这个方法如此之精妙，我从梦中醒来后兴奋的睡不着。技巧的核心是用不同的比特位来表示上面的逻辑描述（或者叫约束），基本上这个技巧可以用在很多逻辑推理的问题中：<br>

| 房子位置 | 房子颜色 | 国籍 | 抽的烟 | 喝的饮料 | 养的宠物 | 隔壁或左或右 |
|:------|:------|:------|:------|:------|:------:|:------|
|01010|01010|01010|01010|01010|01010|010|

房子位置：从左往右，第n位如果为1表示第n+1间房，例如01000表示第二间房，10000表示第一间房，以此类推。<br>
房子颜色：从左往右，依次表示红绿黄蓝白<br>
国籍：从左往右，依次表示英瑞丹挪德<br>
烟：从左往右，依次表示PM、DH、BL、BM、PR牌子<br>
饮料:从左往右，依次表示T、C、M、B、W<br>
宠物:从左往右，依次表示D、B、C、H、F<br>
隔壁或左右：从左往右，依次表示隔壁、左、右<br>
所以改写上面15个约束为一个多项式方程$y=f(x) $ 的约束，它满足下面的要求：

| 约束| 房子位置 | 房子颜色 | 国籍 | 抽的烟 | 喝的饮料 | 养的宠物 | 隔壁或左或右 | $y=f(x) $|

|1|00000|01000|10000|00000|00000|00000|000| 0b00000 10000 10000 00000 00000 00000 000|
|2|00000|00000|01000|00000|00000|10000|000| 0b00000 00000 01000 00000 00000 10000 000|
|3|00000|00000|00100|00000|10000|00000|000| 0b00000 00000 00100 00000 10000 00000 000|
|4|00000|01000|00000|00000|00000|00000|010| 0b00000 00001 00000 00000 00000 00000 001|
|5|00000|01000|00000|00000|01000|00000|000| 0b00000 01000 00000 00000 01000 00000 000|
|6|00000|00000|00000|10000|00000|01000|000| 0b00000 00000 00000 10000 00000 01000 000|
|7|00000|00100|00000|01000|00000|00000|000| 0b00000 00100 00000 01000 00000 00000 000|
|8|00100|00000|00000|00000|00100|00000|000| 0b00100 00000 00000 00000 00100 00000 000|
|9|10000|00000|00010|00000|00000|00000|000| 0b10000 00000 00010 00000 00000 00000 000|

|14|00000|00000|00010|00000|10000|00000|100| 0b00000 00010 00000 00000 00000 00000 100|
|15|00000|00000|00000|00100|00000|00000|000| 0b00000 00000 00000 00001 00000 00000 100|

这个难题的答案是德国人养鱼，翻译成这里的数学表达就是：
$$ f(0b00000 00000 00001 00000 00000 00000 000) = 0b00000 00000 00000 00000 00000 00001 000 $$
或：
$$ f(262144) = 8 $$
亦即Prover知道方程 $f(x) = 8 $有一个解，$ x = 262144 $
