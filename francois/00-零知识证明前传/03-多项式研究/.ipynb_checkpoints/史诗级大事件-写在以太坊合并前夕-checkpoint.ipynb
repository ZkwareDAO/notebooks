# 史诗级大事件-写在以太坊合并前夕

@zkware 2022/08/31

## 以太坊合并路线图



合并分为两个步骤，第一步是在某个 epoch 高度时，在共识层触发的 Bellatrix 网络升级。随后，执行层从工作量证明（PoW）过渡到权益证明（PoS），这个步骤被称为 Paris，它是在 TTD 总难度值达到预定值后触发。

Bellatrix 升级计划于 2022年9月6日 UTC 时间上午 11:34:47 在信标链高度达到 144896 时进行。

而执行层升级 Paris，将在 TTD 总难度值达到 5875000000000000 时触发，预计将在 2022年9月10日-9月20日之间。

## 合并的直接影响

转向权益证明将对以太坊的可持续性产生重大影响，估计能耗将减少 99.95%。从环境，社会和治理（ESG）的角度来看，这显然是一个很大的改进。

工作量证明和权益证明是共识机制，使计算机网络能够就其当前状态达成一致。在这种情况下，计算机网络共同保护和维护区块链（它实际上只是一个事务数据库）。简而言之，共识机制使区块链能够以分布式方式安全运行。

99.95% 的能源减少将来自以太坊从 PoW 的过渡，这是一个保护网络的的过程，主要成本是电力和硬件。在 PoS 系统中，保护网络的主要成本是质押代币形式的资本，并且根本不需要太多的电力。

## 以太坊的供应将进入通缩状态

促进我们今天 DeFi 交易的以太坊是执行层，其在工作量证明上运行。信标链共识层利用权益证明。两者的合并是以太坊过渡到权益证明之时。目前，以太坊的新发行量约为 14600 ETH /天，这是来自主网上挖矿奖励的 13000 ETH 和来自信标链上质押奖励的 1600 ETH 的总和。在“合并”之后，将没有工作量证明，因此没有挖矿奖励，每天只留下 1600 ETH 的质押奖励。

一年前，London upgrade上线，为每笔被视为有效的交易引入了最低费用（称为基本费用）。根据以太坊网站的说法，这笔费用随后被烧毁，基于16 gwei 的汽油平均价格，每天总供应量中少掉了大约 1600 ETH。（一个 gwei 是一个 ETH 的十亿分之一。)

在“合并”之后，以太坊的发行率为 1600 ETH /天的质押奖励减去费用，净增为零。再减去验证者产生的罚款（例如，因行为不端而被削减）和随着时间的推移而丢失的 ETH，这将使以太坊发行有可能变成净负值。

## 以太坊有可能成为加密国度的国债

在LIDO、Coinbase等中心化和去中心化的领先质押者，通过提供POS共识验证服务，而获得以太坊网络的奖励，目前按LIDO实际跑的数据，收益大概在APR 5%左右。（简单计算也支持这个结论：按每天信标链的奖励1600计算，一年58.4万个，目前有1050 万个 ETH 质押在信标链中，以 ETH 币本位计算的收益率达到了可观的 5.5%。） <br>

一旦合并发生后，矿工还将赚取手续费，其收益率将从 5% 上升到 ~10-15%。回报增加了 2-3 倍。目前全部的eth发行量接近1.2亿个，如此高的收益率，将会刺激大量的eth主动锁仓赚取质押挖矿的收益，预计最少有3000万eth进入质押挖矿的市场。如此同时，过去每天大量的PoW矿工常规抛售手上的eth用以支付电费人工等开支的情况将不复存在。<br>

在不考虑币价波动的前提下，一个有稳定收益回报的产品，同时其供应总量将保持一定的恒定，不用担心超发通胀，对于很多长期资金来说，这是一个可以媲美国债的投资标的，毕竟我们最近几年“活久见”了太多的主权货币一夜贬值50%以上的案例。有分析报告已经指出，在一些机构看来，把购买和质押以太坊当作投资一个新兴国家的国债的风险等级是相当的。

## 对加密货币整体的影响

Amber Group在3月份的时候给出了一个非常乐观的分析，认为以太坊的此次合并升级相当于比特币的三重减半，众所周知每次比特币的减半周期都会带来一波牛市。


按照以太坊总市值和比特币的比例关系，Amber认为最少也有一个比特币的一重减半的拉动效果，然后，嗯，2024年来了，新的比特币减半周期到了。所以Amber认为以太坊合并可能会带领加密货币市场走出熊市。

## 反对的声音很多很多


首先是以太坊矿工，按当前的价格1600美元计算，每年PoW矿工可以获得90亿+美金的收入，按现在平均18个月回本周期计算，他们投入了最少120亿美金的设备，所以不光是收入没有了，全市场超过100亿美金的设备怎么处理也是一个大问题。<br>
其次是监管机构，美国政府2022年8月8日宣布制裁tornadocash，在crypto世界里引发轩然大波，大家都在讨论，当以太坊从PoW共识升级为PoS共识后，监管就可能很容易影响PoS节点，要求他们制裁某些地址的交易等等，这在以太坊社区里被视为一种攻击，但确实也暴露出来PoS相对于PoW的去中心化要低一个级别的事实。这个来自监管的压力，同时也对crypto的未来蒙上了一层阴影。 <br>
还有一个因素，时间，因为9月份对于crypto真是一个不堪回首的月份，17年的9.4，21年的9.24，对韭菜都是当头棒喝，所以很多人选择在9月份看空。<br>
最后是宏观分析者，他们认为crypto已经完全和美股同步，在美联储的强加息预期下，所有风险资产只有一个下场，那就是熊市。
