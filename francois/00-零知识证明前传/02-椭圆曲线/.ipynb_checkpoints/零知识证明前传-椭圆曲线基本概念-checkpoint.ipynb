## 椭圆曲线

首先要说明的一点是，椭圆曲线不是椭圆。椭圆方程是下面这样的：

$$ E = \{(x,y) \in R^2 |\frac{x^2}{a^2}+\frac{y^2}{b^2} =1 \} $$

而通常我们讨论的椭圆曲线的曲线方程是一个二元三次方程，它有多种形式，在椭圆曲线密码体系中，最常用的是如下的Weierstrass通用式：
$$ E = \{(x,y) \in R^2 |y^2 = x^3+ax+b, 4a^3+27b^2 \neq 0  \} $$
之所以取名叫椭圆曲线，是因为该曲线方程跟求椭圆弧长的积分公式相似。从曲线方程和图像易知，椭圆曲线关于X轴对称。判定式不等于零是为了椭圆曲线不存在奇异点，即处处光滑可导，这样才能进行椭圆曲线上的加法运算。

## 实数域椭圆曲线加法群

无穷远点：定义一个无穷远点O，即经过椭圆上任意一点的与X轴垂直的直线都经过该点。（在非欧几何中，可认为平行线在无穷远处会交于一点。）

椭圆曲线点加法：椭圆曲线上经过 P 和 Q 两个点的直线与椭圆曲线的交点记作 $R=P<>Q$ ，根据定义有 $P<>Q=Q<>P$ 以及$O<>(O<>P)=P$ 。继而定义椭圆曲线点加法：$P+Q=O<>(P<>Q)$，即加法结果是经过点R 且与 X 轴垂直的直线与椭圆曲线的另外一个交点-R，简单来说，就是交点 R 关于 X 轴的对称点-R。几何意义如下：


椭圆曲线加法群：定义为椭圆曲线在实数域上的点集以及点加法 <br>
封闭性：元素是椭圆曲线上的点，且根据加法定义，加法运算得到的点都在椭圆曲线上，满足封闭性。<br>
单位元：选取无穷远点作为单位元，记为O ，易证得 $O+P=P$。<br>
逆元：因为椭圆曲线过于X轴对称，$P(x,y)$ 的关于 X 轴对称的点$P(x, -y)$ 就是逆元。因为按照O的定义$O=P+(-P)$ 。<br>
结合律：即验证$P+(Q+R)=(P+Q)+R$，通过几何作图可以验证得到的两个点确实相同，更严格的证明需要用到 [Cayley–Bacharach theorem]，即假设两条三次曲线有9个交点，如果第三条三次曲线经过前两条三次曲线的8个交点，那么它也必定通过第9个交点。<br>


(P+Q)+R

P+(Q+R)

由此可知，椭圆曲线上的点在椭圆曲线加法运算上构成了一个加法群。<br>
又根据定义可以直接证明交换律是成立的，因此椭圆曲线上的点在椭圆曲线加法运算上构成了一个阿贝尔群。增加了单位元后，椭圆曲线方程改为：

$$E = \{(x,y) \in R^2 |y^2 = x^3+ax+b, 4a^3+27b^2 \neq 0  \} \cup \{O\} $$

### 代数运算
上文从几何意义上给出了椭圆曲线点加法，在实际计算的时候需要用方程组计算。要注意的是，这并不是两个点的坐标简单相加。
$$ y^2 = x^3+ax+b $$
$$ y=mx+c $$
有中学的数学基础，就可以计算出来。而根据经过两个不同点$P(x_p,y_p), Q(x_q,y_q)$的直线，用拉格朗日插值法，或者中学数学知识都可以得到直线的系数表达，如果P和Q相同，则m为切线斜率，即求导数，此处就不多加阐述。得到R的坐标为：$R(m^2-x_p-x_q,  y_p+m(x_p-x_q))$ ,而真正的加法结果是 $-R(m^2-x_p-x_q,  -y_p-m(x_p-x_q))$。

## 有限域椭圆曲线

在密码学里，我们通常只对有限域$\mathbb{F}_p $上的椭圆曲线感兴趣，将上面的方程改写为：$ E = \{(x,y) \in F_q |y^2 = x^3+ax+b, 4a^3+27b^2 \neq 0  \} $，这里的加法、乘法都是$\mathbb{F}_p $上的模加和模乘法，在有限域上下文中，加法和乘法都是模p运算。在上面有限域椭圆曲线定义下，所有的(x,y)取值将构成一个集合，可以发现，椭圆曲线变成了离散的点，且关于 $y = p/2$ 对称



## 有限域椭圆曲线加法群

在这个集合上如果定义一个加法如下：$P+Q=-R$，其中 $-R(m^2-x_p-x_q,  -y_p-m(x_p-x_q))$，所有的运算都是模P的加法或乘法。椭圆曲线在素数域 $\mathbb{F}_p$ 上的点加法依然构成阿贝尔群。单位元依旧是无穷远点，元素$P(x_p, y_p)$的逆元变成 $-P(x_p, -y_p \mod p)$ 。而交换律、结合律、封闭性则可以通过素域上的模加法、模乘法来证明。实数域的椭圆曲线点加法定义是有明确几何意义的，从几何上好证明。而椭圆曲线在 $\mathbb{F}_p$ 就没有明显的几何意义了，观察可发现 P,Q,R 三点满足 $ax+by+c=0 \mod p$。

这个加法群的点是无序无规则的，很难直接应用，所以在密码学里一般对其循环子群来操作，因为循环子群可以通过生成元来定义，加上其良好的循环特性，非常适合密码学的应用场景。 <br>
在有限域下，椭圆曲线加法群的元素是有限的，元素数目就是群的阶。Schoof算法运用了 Hasses 定理。Hasses定理给出了椭圆曲线在 $\mathbb{F}_p $ 的阶的范围，当 p 很大时，阶跟 p 的值是比较接近的。

### 有限域椭圆曲线加法群的循环子群

P 的标量乘法得到的点集$nP$是原椭圆曲线群的一个子集，则它是原椭圆曲线群的一个子群，而且是循环子群。子群中元素个数称为子群的阶，点 P 称为该子群的基点或者生成元。循环子群是椭圆曲线密码体系的基础，我们期望子群中元素越多越好，如何找到一个合适的子群尤为重要。根据拉格朗日的群论定理，子群的阶n是父群的阶N的约数。即有 $h=N/n$，h 是一个整数，我们称之为子群的余因子(cofactor)。如果$h=1$，意味着子群的阶等于父群的阶，相比于直接用父群，我们用到了循环子群的特点，即增加了一个生成元G用来更好的刻画这个群的特征。

### 有限域椭圆曲线加法群的循环子群的密码体系

有了上面的定义，可以对有限域椭圆曲线加法群的循环子群$\{i*G|i \in Z_{n}, G为生成元,n为子群的阶 \}$上的加密做一个定义，任意给一个秘密x，要求$x \in \mathbb{F}_p $，给定生成元G, 则密文就是 $x*G$。可以看到，明文是一个任意的整数，而密文则是一个椭圆曲线上的点。仅仅知道$x*G$ 和 G是无法求出来x的，这一点由ECDLP来保证，即是所谓的椭圆曲线离散对数难题。

总结起来，椭圆曲线加密算法工作在素数域下的椭圆曲线循环子群中，需要的域参数(Domain Parameter)包括 ：

p: 素数域的大小。<br>
a, b: 椭圆曲线 $y^2 = x^3+ax+b$ 的系数。<br>
G：生成子群的基点。<br>
n：子群的阶。<br>
h：子群的余因子。<br>
如比特币用来做数字签名中采用的椭圆曲线 [secp256k1] 的域参数如下：<br>

p = 0xFFFFFFFF FFFFFFFF FFFFFFFF FFFFFFFF FFFFFFFF FFFFFFFF FFFFFFFE FFFFFC2F = $2^256-2^32-2^9-2^8-2^7-2^6-2^4-1$ <br> 
a = 0, b = 7，即曲线方程是 $y^2 = x^3+7$ 。 <br>
G = (0x79BE667E F9DCBBAC 55A06295 CE870B07 029BFCDB 2DCE28D9 59F2815B 16F81798,
0x483ADA77 26A3C465 5DA4FBFC 0E1108A8 FD17B448 A6855419 9C47D08F FB10D4B8)<br>
n = 0xFFFFFFFF FFFFFFFF FFFFFFFF FFFFFFFE BAAEDCE6 AF48A03B BFD25E8C D0364141<br>
h = 1<br>

利用sage可以进一步研究下上面几个参数的意义，可以看到父群的阶$N=E.order()$是等于子群的阶n的(h=1)，并且这个阶n也是一个很大的素数，所以这个曲线是一个精心选择的曲线，里面的参数如何计算出来，这是一个迷。<br>


### 比特币的公钥和地址
通常用戶⻅到的比特币地址是经过“Base58Check”编码的,这种编码使用 了58个字符(一种Base58数字系统)和校验码，提高了可读性、避免歧义并有效防止了在地址转录和输入中产生的错 误。Base58Check编码也被用于比特币的其它地方，例如比特币地址、私钥、加密的密钥和脚本哈希中，用来提高可读性和 录入的正确性。
